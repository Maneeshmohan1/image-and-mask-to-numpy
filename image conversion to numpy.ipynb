```python
import os
import random
import numpy as np
import matplotlib.pyplot as plt
from itertools import chain
from skimage.io import imread, imshow, concatenate_images
from skimage.transform import resize
import cv2
```

```python
#Dataset path
DATA_PATH = 'C:/Users/shyam/Desktop/monuseg_CIA/Monuseg/monuseg/TrainData/'
```

```python
#path of the training images 
train_path = DATA_PATH + 'original_images/'
#path of the mask images
mask_path = DATA_PATH  + 'Bin/'
```

```python
train_dir = sorted(os.listdir(train_path))
mask_dir = sorted(os.listdir(mask_path))
```

```python
count = 1
#output folder
out = "C:/Users/shyam/Desktop/monuseg_CIA/npy/"
for x in range(len(train_dir)):
    image = train_dir[x].split('.')[0]
    img = cv2.imread(train_path + train_dir[x]) 
    print(x)
    print('Preprocessing: ' + image + ': ' + str(count))
    count += 1
    X = np.zeros((1000, 1000, 4), dtype=np.uint8)
    y = cv2.imread(mask_path + mask_dir[x])
    y = resize(y, (1000, 1000, 1), mode='reflect', preserve_range=True)
    X[...,:3] = img
    X[...,3:] = y
    np.save(out+image, X)
    
```
